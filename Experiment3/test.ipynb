### 导入图片

```python
from PIL import Image
from pylab import *

axis('off')
pil_im = Image.open("test.jpg")
im = array(pil_im)
imshow(im);
```

### 调整灰度

```python
grey = pil_im.convert("L")
im = array(grey)
imshow(im)
axis('off')
```

### 灰度变换

```python
im = array(pil_im)
pic = 255.0 * (im/255.0)**2
imshow(pic)
axis('off')
```

### 复制，反转

```python
region = (1000, 0, 3500, 2000)
region_rev = pil_im.crop(region)
region_rev = region_rev.transpose(Image.ROTATE_180)
imshow(array(region_rev))
axis('off')
```

### 图像轮廓

```python
figure()
gray()
contour(im, origin='image')
axis('equal')
axis('off')
```

### 图像模糊化

```python
from numpy import *
from scipy.ndimage import filters
im = array(pil_im)
for i in range(3):
    im[:, :, i] = filters.gaussian_filter(im[:, :, i], 10)
imshow(im)
axis('off')
```
